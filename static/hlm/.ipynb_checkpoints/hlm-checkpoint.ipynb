```python
import bs4, requests, re
from bs4.element import Tag
```

```python
def browse(url, protocol='https://', encoding='utf-8'):
    return bs4.BeautifulSoup(requests.get(protocol + url).content.decode(encoding))
```

```python
soup = browse('https://www.zhonghuadiancang.com/wenxueyishu/hongloumeng').find_all('a')
```

```python
data = [(i.string.replace('\u3000', ' '), i['href']) for i in soup if '回' in i.get_text()]
```

```python
for i, (title, url) in enumerate(data):
    text = requests.get(url).content.decode().replace('\r', '')
    soup = bs4.BeautifulSoup(text)
    open('{}.html'.format(i), 'w', encoding='utf-8').write(
        '<div style="font-size: 24px">\n' +
        '\n'.join(map(str, soup.find_all('p')[1:-1])) +
        '\n</div>'
    )
    print(title)
```

```python
open('index.html', 'w', encoding='utf-8').write('\n'.join(
    '<a href="{}.html">{}</a><br>'.format(i, title.strip())
    for i, (title, _) in enumerate(data)
))
```

```python
from selenium import webdriver
options = webdriver.EdgeOptions()
options.add_argument('--headless')
browser = webdriver.Edge(options)
```

```python
browser.get('https://www.99csw.com/book/8489/300147.htm')
```

```python
browser.page_source
```

```python
chapters = []
for i, link in enumerate(links):
    browser.get('https://www.99csw.com' + link['href'])
    soup = bs4.BeautifulSoup(browser.page_source)
    chapters.append(soup.find('div', id='content'))
```

```python
chapters
```

```python
for i, chapter in enumerate(chapters):
    open('{}.html'.format(i), 'w', encoding='utf-8').write(
        str(chapter.find('h2')) + '\n' +
        '\n'.join(
            '<p>{}</p>'.format(re.sub(r'九.*?九.*?藏.*?书.*?网|(?:www\.)99lib.net', '', div.get_text()))
            for div in chapter.find_all('div')
        )
    )
    print(chapter.find('h2').string)
```
